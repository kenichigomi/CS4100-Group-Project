```python
import osmnx as ox
import networkx as nx

import matplotlib.pyplot as plt
```

```python
start_point = (42.3601, -71.0589)
G = ox.graph_from_point(start_point, dist=1000, network_type='walk')
origin_node = ox.nearest_nodes(G, Y=start_point[0], X=start_point[1])
```

```python
target_distance = 10 # in meters
nodes_within_x = nx.single_source_shortest_path_length(G, origin_node, cutoff=target_distance)
```

```python
nx.has_path(G, origin_node, temp)
```

```python
length = nx.shortest_path_length(G, origin_node, temp, weight="length")
length
```

```python
test_paths = ox.k_shortest_path(G, origin_node, temp, weight="length")
```

```python
route = nx.shortest_path(G, origin_node, temp, weight="length")
```

```python
fig, ax = ox.plot_graph_route(G, route, route_color='red', route_linewidth=4, node_size=10, show=True, close=True)
```
